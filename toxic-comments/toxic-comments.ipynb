```python
import sys, os, re, csv, codecs, numpy as np, pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.layers import Dense, Input, LSTM, Embedding, Dropout, Activation
from tensorflow.keras.layers import Bidirectional, GlobalMaxPool1D
from tensorflow.keras.models import Model
from tensorflow.keras import initializers, regularizers, constraints, optimizers, layers
```

```python
train = pd.read_csv('train.csv')
sub = pd.read_csv('sample_submission.csv')
test = pd.read_csv('test.csv')
```

```python
comments_train = train['comment_text']
comments_test = test['comment_text']
```

```python
train.shape
```

```python
train.isna().sum()
```

```python
y = train.iloc[:,2:9]
```

```python
max_features = 20000
tokenizer = Tokenizer(num_words=max_features)
tokenizer.fit_on_texts(list(comments_train))
list_tokenized_train = tokenizer.texts_to_sequences(comments_train)
list_tokenized_test = tokenizer.texts_to_sequences(comments_test)
```

```python
# get occurances of words
tokenizer.word_counts
```

```python
# get indexes of words
tokenizer.word_index
```

```python
list_tokenized_train[:1] # index fist row (0 to 1)
```

```python
# need to make vectors the same size
maxlen = 200
X_train = pad_sequences(list_tokenized_train, maxlen=maxlen)
X_test = pad_sequences(list_tokenized_test, maxlen=maxlen)
```

```python
# how do we know to use 200 as max length of vectors?
# see distribution of number of words in comments
total_wordcount = [len(one_comment) for one_comment in list_tokenized_train]
```

```python
# we see that typical sentence length is ~30
plt.hist(total_wordcount, bins = np.arange(0,410,10))
```

```python
# define input layer
inp = Input(shape=(maxlen,))
```

```python
# pass onto embedding layer
# project words onto vector space
embed_size = 128 # arbitrary
x = Embedding(max_features,embed_size)(inp)
```

```python
x = LSTM(60, return_sequences=True,name='lstm_layer')(x)
```

```python
# reshape 3d tensor to 2d
# to avoid throwing away important data
x = GlobalMaxPool1D()(x)
```

```python
# disable a portion of nodes to force nodes in next layer to handle representation of missing data
# for better representation
# set to drop out 10% of the nodes
x = Dropout(0.1)(x)
```

```python
# connect output of drop out layer to densely connected layer
# output passes through RELU function
# Activation((Input*Weights)+Biases)
x = Dense(50, activation='relu')(x)
```

```python
x = Dropout(0.1)(x)
```

```python
# feed next output through sigmoid function
# b/c we want binary classification for each of the 6 labels 
# sigmoid function will provide values from 0 to 1
x = Dense(6, activation = 'sigmoid')(x)
```

```python
# define outputs, inputs, and configure learning process
# define loss function to be 'binary_crossentropy'
model = Model(inputs=inp,outputs=x)
model.compile(loss='binary_crossentropy',
             optimizer='adam',
             metrics=['accuracy'])
```

```python
# test model
# 10 validation set
batch_size = 32 # arbitrary
epochs = 2 # arbitrary
model.fit(X_train, y, batch_size = batch_size, epochs = epochs, validation_split = 0.1)
```

```python
# get summary
model.summary()
```

```python
# check output performance
from tensorflow.keras import backend as K

# with Sequential model
get_3rd_layer_output = K.function([model.layers[0].input],
                                  [model.layers[2].output])
layer_output = get_3rd_layer_output([X_train[:1]])[0]
layer_output.shape
```

```python
pred = model.predict(X_test)
```

```python
data = pd.DataFrame(data=pred,columns=["toxic", "severe_toxic", "obscene", "threat", "insult", "identity_hate"])
```

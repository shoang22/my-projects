```python
X = df.drop('benign_0__mal_1',axis=1).values
y = df['benign_0__mal_1'].values
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=.25, random_state=101)
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
```

```python
X_train = scaler.fit_transform(X_train)
```

```python
X_test = scaler.transform(X_test)
```

```python
from tensorflow.keras.models import Sequential
```

```python
from tensorflow.keras.layers import Dense,Dropout
```

```python
X_train.shape
```

```python
model = Sequential()
model.add(Dense(30,activation='relu'))
model.add(Dense(15,activation='relu'))
model.add(Dense(1,activation='sigmoid')) # because of binary classification problem
model.compile(loss='binary_crossentropy',optimizer='adam')
```

```python
model.fit(x=X_train,y=y_train,epochs=600,validation_data=(X_test,y_test))
```

```python
losses = pd.DataFrame(model.history.history)
```

```python
losses.plot() # validation loss is increasing -> overfitting
```

```python
# Early Stopping with TF.Keras Callback
# Dropout -> fraction of which to turn off neurons during each batch
model = Sequential()

model.add(Dense(30,activation='relu'))
model.add(Dropout(rate=.5)) # half neurons (randomly selected) will be turned off

model.add(Dense(15,activation='relu'))
model.add(Dropout(rate=.5))

model.add(Dense(1,activation='sigmoid')) # because of binary classification problem
model.compile(loss='binary_crossentropy',optimizer='adam')
```

```python
from tensorflow.keras.callbacks import EarlyStopping
```

```python
help(EarlyStopping)
```

```python
early_stop = EarlyStopping(monitor='val_loss',verbose=1,patience=25)
```

```python
model.fit(x=X_train,y=y_train,epochs=600,validation_data=(X_test,y_test),callbacks=[early_stop])
```

```python
model_loss=pd.DataFrame(model.history.history)
```

```python
model_loss.plot()
```

```python
predictions = model.predict_classes(X_test)
```

```python
from sklearn.metrics import classification_report,confusion_matrix
```

```python
print(classification_report(y_test,predictions))
```

```python
print(confusion_matrix(y_test,predictions))
```
